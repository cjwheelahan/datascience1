So the two datasets look very similar, although there appears to be a bit more noise in the first radar observations. What I'd like to do is to append the two datasets (that is, use both), and create a weighting function whereby the radar2 observations will get more weight than the radar 1 observations in predicting X, Y and Z values as a function of time. In order to create that weighting function, let's make sure a linear fit is best, then see how noisy the datasets actually are.

```python
def get_best_model(depen, feat):
    results = []
    for i in range(3):
        poly = feat**(i+1)
        polymodel = OLS(depen, poly)
        ols_res = polymodel.fit()
        results.append(ols_res.rsquared)

    print "The best model is a", results.index(max(results))+1, "th degree polynomial, with an R-squared of", max(results)

print "For Radar 1 X, Y and Z:"
get_best_model(radar1t, radar1x)
get_best_model(radar1t, radar1y)
get_best_model(radar1t, radar1z)

print "For Radar 2 X, Y and Z:"
get_best_model(radar2t, radar2x)
get_best_model(radar2t, radar2y)
get_best_model(radar2t, radar2z)
```

We now know that a first degree polynomial fit is best for all of our axes. Interestingly, both radars have a very high R-Squared, so the Radar 1 dataset might be less troublesome than expected. Let's try to create a weighting function anyway, just to be thorough.

What we'll do is see the relative proportions of variance explained by Radar 1 and Radar 2. The amount worse-off Radar 1 as a proportion of Radar 2 will be our weight.

```python
def rsq(time, axis):
    model = OLS(axis, time)
    results = model.fit()
    return results.rsquared

r1x = rsq(radar1t, radar1x)
r2x = rsq(radar2t, radar2x)
r1y = rsq(radar1t, radar1y)
r2y = rsq(radar2t, radar2y)
r1z = rsq(radar1t, radar1z)
r2z = rsq(radar2t, radar2z)

# Print the R-squared values and the weighting value for Radar 1
print "For Radar 1 x, the R-Squared is:", r1x
print "For Radar 2 x, the R-Squared is:", r2x
print "A weighting factor for Radar 1 X is:", r1x/r2x

print "For Radar 1 y, the R-Squared is:", r1y
print "For Radar 2 y, the R-Squared is:", r2y
print "A weighting factor for Radar 1 Y is:", r1y/r2y

print "For Radar 1 time, the R-Squared is:", r1z
print "For Radar 2 time, the R-Squared is:", r2z
print "A weighting factor for Radar 1 time is:", r1z/r2z
```

Putting it all together...

```python
StWdf = pd.DataFrame(data=StW[:, :])
StWdf = StWdf[:-10]
radar2 = radar2.loc[:, 1:]
lists = [StWdf, radar2]
bang = pd.concat(lists, axis = 1)
bang.columns = ['time', 'radar 1 x', 'radar 1 y', 'radar 1 z', 'radar 2 x', 'radar 2 y', 'radar 2 z']

# weighted averages for X, Y and Z as a function of time
x = (bang['radar 1 x']*(r1x/r2x) + bang['radar 2 x']) / ((r1x/r2x)+1)
y = (bang['radar 1 y']*(r1y/r2y) + bang['radar 2 y']) / ((r1y/r2y)+1)
z = (bang['radar 1 z']*(r1z/r2z) + bang['radar 2 z']) / ((r1z/r2z)+1)
```

## Making projections (90% confidence intervals of X and Y) using Bootstrapping

Now that we have our weighted X, Y and Z as a function of time, what we want to do is create a predictive model that projects X and Y when Z is 0 (when the meteorite hits the ground). In other words, we want to see what the intercepts are when regressing Z on X and Y.

```python
def project_CI(coord, z):
    slope = []
    inter = []
    n = coord.shape[0]
    sub_size = n/10

    for i in range(250):
        perm = np.random.permutation(n) # Generate a list of indices 0 to n and permute it

        coord_train = coord.iloc[perm[:sub_size]] # Get x-vals for the first n/10 indices in permuted list
        z_train = z.iloc[perm[:sub_size]]

        z2 = sm.add_constant(z_train)
        x_model = OLS(coord_train, z2)
        x_ols = x_model.fit()
        inter.append(x_ols.params[0])
        slope.append(x_ols.params[1])

    conf_int_left = np.percentile(inter, 5)
    conf_int_right = np.percentile(inter, 95)

    print "Mean projection at Z=0 is:", np.mean(inter)
    print "Left Confidence interval at Z=0 is:", conf_int_left
    print "Right Confidence interval at Z=0 is:", conf_int_right
    return conf_int_left, conf_int_right, np.mean(inter), inter

def plot_CI(left, right, mean, inter):
    coord_fig = plt.figure(figsize=(16, 16))
    xvz = coord_fig.add_subplot(221)
    xvz.hist(inter, alpha=0.5)

    # Plot vertical lines at mean and left, right extremes of confidence interval
    xvz.axvline(x = mean, linewidth=3)
    xvz.axvline(x = left, linewidth=1, c='r')
    xvz.axvline(x = right, linewidth=1, c='r')

    plt.title('Histogram of predictions at Z=0')
    plt.ylabel('Count')
    plt.xlabel('Coordinate')
    
    #plt.legend()
```

```python
print "The projections for x coordinates when Z=0 are:"
x_left, x_right, x_mean, x_inter = project_CI(x, z)

print "The projections for y coordinates when Z=0 are:"
y_left, y_right, y_mean, y_inter = project_CI(y, z)

plot_CI(x_left, x_right, x_mean, x_inter)
plot_CI(y_left, y_right, y_mean, y_inter)
```

# Impacted Population Estimations

Now that we have our projections, we can start to evacuate folks. Let's see how many people fall within the X and Y confidence intervals.

```python
# create new df with only the at-risk populations (those inside our confidence intervals)
risk_pop = pop_data[(pop_data.x >= x_left) 
                    & (pop_data.x <= x_right) 
                    & (pop_data.y >= y_left) 
                    & (pop_data.y <= y_right)]

print "The number of residents potentially impacted is", np.sum(risk_pop.residents), "."

#sanity_check(pop_data)
#sanity_check(risk_pop)
```
